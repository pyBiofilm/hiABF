```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from os import path
from sklearn import metrics
from sklearn.metrics import accuracy_score, cohen_kappa_score, roc_auc_score, matthews_corrcoef, average_precision_score,confusion_matrix,precision_score,f1_score,recall_score
from sklearn.metrics import classification_report
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
import seaborn as sns
from sklearn.feature_selection import RFECV
```

```python
## Best on binary classifiers performance, six models were selected as best models
top_models = []
top_models.append(('logreg', LogisticRegression(fit_intercept=True)))
top_models.append(('rfc', RandomForestClassifier(random_state=seed)))
# top_models.append(('svc_poly', SVC(kernel="poly",degree=3,probability=(True))))
top_models.append(('svc_lnr', SVC(kernel="linear", C=1.0,probability=True)))
top_models.append(('adc', AdaBoostClassifier(n_estimators=50)))
# top_models.append(('svc_rbf', SVC(probability=True)))
```

```python
results = []
names = []
seed= 42
kfold = KFold(n_splits=10, random_state=seed, shuffle= True)

for name, model in top_models:

    # Apply RFECV
    rfecv_model = RFECV(model, step=1, cv=kfold)
    rfecv = rfecv_model.fit(x_train, y_train)
    model_train_RFE = rfecv.transform(x_train)
    model_test_RFE = rfecv.transform(x_test)
    # Scores means and std deviations
    cv_scores = cross_val_score(model, model_train_RFE, y_train, cv=kfold, scoring='accuracy')
    results.append(cv_scores)
    names.append(name)
    print("%s: Accuracy Score %0.2f%% (%0.2f%%)" % (name, 100*cv_scores.mean(), 100*cv_scores.std()))
    print("Optimal number of features : %d" % (rfecv.n_features_))
    print("")
    
    # Predictions train set
    cv_preds = cross_val_predict(model, model_train_RFE, y_train,cv=kfold)
    #print("%s: Accuracy %0.2f%%" % (name, 100*metrics.accuracy_score(y_train, cv_preds)))
    print("%s: Precision-Recall %0.2f%%" % (name, 100*metrics.average_precision_score(y_train, cv_preds)))
    print("%s: Matthews Coefficient %0.2f%%" % (name, 100*matthews_corrcoef(y_train, cv_preds)))
    print("%s: Cohen Kappa Score %0.2f%%" % (name, 100*metrics.cohen_kappa_score(y_train, cv_preds)))
    print("%s: ROC AUC Score %0.2f%%" % (name, 100*metrics.roc_auc_score(y_train, cv_preds)))
    print("")
    target_names = ['nonABF', 'ABF']
    print(classification_report(y_train, cv_preds, target_names=target_names))
    
    # Predictions Validation Set
    cv_preds2 = cross_val_predict(model, model_test_RFE, y_test, cv=kfold)
    print("%s: Accuracy %0.2f%%" % (name, 100*metrics.accuracy_score(y_test, cv_preds2)))
    print("%s: Precision-Recall %0.2f%%" % (name, 100*metrics.average_precision_score(y_test, cv_preds2)))
    print("%s: Matthews Coefficient %0.2f%%" % (name, 100*matthews_corrcoef(y_test, cv_preds2)))
    print("%s: Cohen Kappa Score %0.2f%%" % (name, 100*metrics.cohen_kappa_score(y_test, cv_preds2)))
    print("%s: ROC AUC Score %0.2f%%" % (name, 100*metrics.roc_auc_score(y_test, cv_preds2)))
    print("")
    target_names = ['nonABF', 'ABF']
    print(classification_report(y_test, cv_preds2, target_names=target_names))
```
